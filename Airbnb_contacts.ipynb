```python
import pandas as pd

df=pd.read_csv('airbnb_contacts.csv')
df
```

```python
new_df=df.iloc[:,[True, False, False, False,False,True]]
new_df
```

```python
group_df=new_df.groupby(['id_guest'])[['n_messages']].sum()
group_df
```

```python
sort=group_df.sort_values(by='n_messages',ascending=False)
sort
```

```python
sort['Rank']=sort.rank(method='dense',ascending=False)

# for method in ['average', 'min', 'max', 'dense', 'first']:
#     sort[f'{method}_rank'] = sort.groupby('id_guest')['n_messages'].rank(method)
# sort

sort
```
