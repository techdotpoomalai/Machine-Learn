# 1. Data Preprocessing

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Churn_Modelling.csv')
dataset
```

```python
data=dataset.drop(['RowNumber',	'CustomerId',	'Surname'],axis=1)
data
```

```python
df=data.copy()
df
```

```python
df['Geography'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
label_encoder=LabelEncoder()
hot_encoder=OneHotEncoder()
df['Gender']=label_encoder.fit_transform(df['Gender'])
temp=pd.get_dummies(df['Geography'])
df.drop(['Geography'],axis=1,inplace=True)
fdf=pd.concat([temp, df], axis=1)
fdf=pd.DataFrame(fdf.iloc[:, :].values)
fdf
```

```python
X=fdf.iloc[:, :12]
y=fdf.iloc[:, 12:]
type(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
type(X_train)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train =sc.fit_transform(X_train).astype(np.float32)
X_test =sc.transform(X_test).astype(np.float32)

print(X_test.shape)
print('\n')
print(X_train.shape)
print()
print(type(X_train))
print()
print(X_train)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
print(type(X_train), type(X_test), type(y_train),type(y_test))
```

```python
y_train=np.array(y_train)
y_test=np.array(y_test)
print(type(y_train),type(y_test))
```

```python

X_train = np.array(X_train)
y_train = np.array(y_train)
```

```python

X_train = X_train.astype('float32')
y_train = y_train.astype('int')
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense

classifier = Sequential()
classifier.add(Dense(units=32, activation='relu', kernel_initializer='uniform'))
classifier.add(Dense(units= 6, kernel_initializer='uniform', activation = 'relu'))
classifier.add(Dense(units = 6, kernel_initializer='uniform', activation = 'relu'))
classifier.add(Dense(units= 1,kernel_initializer='uniform', activation = 'sigmoid'))


classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])


classifier.fit(X_train, y_train, batch_size=2, epochs=10)
```
